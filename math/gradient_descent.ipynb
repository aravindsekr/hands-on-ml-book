# Gradient descent
Lets consider a simple model linear regression to under gradient descent. Linear regression is a simple model that will be useful when we need single value as output or discrete value. The way linear regression works is by fitting a straight line that captures most of the sample / training data. In other words, straight line cannont pass through each data point due to fact in real world, almost all data is non-linear. Because of this non-linearity there will be "residual" values for each data point w.r.t to the straight line we fitted. Usually we have different techniques to understand the effectivness of LR straight line like MSE, RMSE, MAE etc. Most of the scenarios we would use MSE or RMSE. Again we will MSE for our example, since it would be easily apply partial derivates to come up with optimal solutions later on. MSE is also best method statistically which we can derive later using Maximum likelikhood method.

Formula for Linear regression line:

$$
\hat{y} = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \dots + \theta_n x_n
$$

In vectorized notation:
$$
\hat{y} = h_\theta(X) = \theta \cdot X
$$

In above formula:  
$\theta$ is Parameter Vector <br/>
$\theta_0$ is bias term <br/>
$\theta_1 \dots \theta_n$ feature weights <br/>
$h_\theta$ is hypothesis function, using model parameters $\theta$ <br/>
$X$ is Feature vector <br/>


Taking dot product would result in straight line, so we have to adjust $\theta$ to come up with line that fits the input vector $X$.

In order to find accuracy of the model, we use MSE which is:

$$
\text{MSE} = \frac{1}{m} \sum_{i=1}^{m} \left( X^{(i)} \theta - y^{(i)} \right)^2
$$

In above formula:
$X$: Vector of size m samples <br/>
$\theta$: Parameter <br/>

Concisely we represent in this way, since $\theta$ would be column vector and $X$ usually will be column vector as well, so we would take transpose of model parameter vector.

$$
\hat{y} = \theta^T X
$$

Rewrite the cost function with parameter $\theta$ we can write like:
$$
MSE(X, h_\theta) = (1/m)\sum_{i=1}^{m} \left(\theta^T x^{(i)} - y^{(i)} \right)^2
$$

Considering Norm form gives this equation:
$$
\|X\theta - y\|^2 = (X\theta - y)^T (X\theta - y)
$$

Then cost function can be written as:
$$
J(\theta) = (1/m)(X\theta - y)^T (X\theta - y)
$$
